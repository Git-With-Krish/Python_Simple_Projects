# PCA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

The Data:
Working with cancer dataset as it has many features

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
cancer.keys()
```

```python
print(cancer['target_names'])  #To get values
```

```python
#Loading data
df = pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
df.head()
```

# PCA Visualization
It is difficult to visualize high dimensional data, we use the PCA to find the first two Principal Components and visualoze the data in new 2D space with single scatter-plot. Before that, we scale our data so that each feature has single unit variance.

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(df)
scaled_data = scaler.transform(df)
```

PCA with Scikit Learn uses a very similar process such as other preprocessing functions that comes from Scikit Learn. We instantiate a PCA object, find the principal components using the fit method, then apply the rotation and dimensionality reduction by calling transform()

We can also specify how many components we want to keep when creating the PCA object

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(scaled_data)
```

Now we can transform this data  into its first 2 principal components

```python
x_pca = pca.transform(scaled_data)
print(scaled_data.shape)
print(x_pca.shape)
```

Let's plot these two dimensions

```python
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0],x_pca[:,1],c=cancer['target'],cmap='plasma')
plt.xlabel("First Component")
plt.ylabel("Second Component")
```

# Interpreting the components
It is difficult to understand what these two components represent. The components corresponds to combination of original features, the components are themselves stored as an attribute of the fitted PCA object.

```python
pca.components_
```

In this numpy array, each row represents a principal component and each column relates back to the original features. We can visialize it in a heat map. 

```python
df_comp = pd.DataFrame(pca.components_,columns=cancer['feature_names'])
plt.figure(figsize=(12,6))
sns.heatmap(df_comp,cmap='plasma')
```

This heatmap and the color bar basically represents the correlations between various feature and the principal component itself.
