# HandCricket Game using Heads or Tails as Toss:

```python
import random  #Inputs for Computer

print("***Welcome to Hand Cricket Player1 vs Computer***")

result = False
while result != True:
    
    #TOSS
    player1_toss = input("Call Heads or Tails: ").upper()
    print("player1 called : ",player1_toss)

    if player1_toss != "H" and player1_toss != "T":
        player1_toss = input("Call Heads or Tails: ").upper()

    toss_choice = ["H", "T"]
    toss = random.choice(toss_choice)
    print("Toss is: ",toss)
    
    val = 0
    
    if player1_toss == toss:
        print("Player1 won the toss and decided to bat first")
        val = 1

    else:
        print("Computer won the toss and decided to bat first")
        val = 2
    
    #GameStarts
    pscore = 0
    cscore = 0
    target = 0
    firstInnings = 0
    secondInnings = 0
    
    if val == 1:
        chaser = "Computer"
        defender = "Player1"
        print("---Player1 Batting---")
        while firstInnings != 1:
            #Player1
            #Value of ranges is 1-6 only
            
            player1_input = int(input("Player1 input is: "))
            
            if player1_input > 6 and player1_input < 1:
                player1_input = int(input("Player1 input is: "))
            
            computer_input = random.randint(1,6)
            print("Computer input is: ",computer_input)
            
            if player1_input != computer_input:
                pscore += player1_input
            else:
                print("---Player1 is OUT---")
                firstInnings = 1
                
        print("{} scored {} runs".format(defender,pscore))
        target = pscore + 1
        print("Target is: ",target)
        
        print("-------------*********END OF AN INNINGS*********-------------")
        
        #Chasing
        print("---{} Batting---".format(chaser))
        wicket ="NO"
        while secondInnings != 1 and wicket != "YES":
        
            if target > cscore:    
                player1_input = int(input("Player1 input is:"))
                
                if player1_input > 6 and player1_input < 1:
                    player1_input = int(input("Player1 input is: "))
                
                computer_input = random.randint(1,6)
                print("Computer input is: ",computer_input)
             
                if player1_input != computer_input and cscore <= pscore:
                    cscore += computer_input
                    if cscore >= pscore:
                        print("----{} won the game..!-----".format(chaser))
                        secondInnings = 1

                else:
                    print("---{} is OUT---".format(chaser))
                    wicket = "YES"
                    secondInnings = 1
        print("{} scored {} runs".format(chaser,cscore))
            
        if wicket == "YES" and pscore > cscore:
            print("-----Player1 won the game..!-----")

        elif wicket == "YES" and cscore > pscore:
            print("-----Computer won the game..!-----")

        elif wicket == "YES" and cscore == pscore:
            print("-----It's a Tie-----")
     
    result = True
    
    if val == 2:
        defender = "Computer"
        chaser = "Player1"
        print("---Computer Batting---")
        
        while firstInnings != 1:
            #Computer Bats
            
            player1_input = int(input("Player1 input is: "))
            
            if player1_input > 6 and player1_input < 1:
                player1_input = int(input("Player1 input is: "))
            
            computer_input = random.randint(1,6)
            print("Computer input is: ",computer_input)
            
            if player1_input != computer_input:
                cscore += computer_input
            else:
                print("---Computer is OUT---")
                firstInnings = 1
        
        print("{} scored {} runs".format(defender,cscore))
        target = cscore + 1
        print("Target is: ",target)
        
        print("-------------*********END OF AN INNINGS*********-------------")
    
        #Chasing    
        print("---{} Batting---".format(chaser))
        wicket ="NO"
        while secondInnings != 1 and wicket != "YES":
        
            if target > pscore:    
                player1_input = int(input("Player1 input is:"))
                
                if player1_input > 6 and player1_input < 1:
                    player1_input = int(input("Player1 input is: "))
                
                computer_input = random.randint(1,6)
                print("Computer input is: ",computer_input)
             
                if player1_input != computer_input and pscore <= cscore:
                    pscore += player1_input
                    if pscore > cscore:
                        print("----{} won the game..!-----".format(chaser))
                        secondInnings = 1

                else:
                    print("---{} is OUT---".format(chaser))
                    wicket = "YES"
                    secondInnings = 1
        print("{} scored {} runs".format(chaser,pscore))
            
        if wicket == "YES" and pscore > cscore:
            print("-----Player1 won the game..!-----")

        elif wicket == "YES" and cscore > pscore:
            print("-----Computer won the game..!-----")

        elif wicket == "YES" and cscore == pscore:
            print("-----It's a Tie-----")

    #Ask User if he/she wanna play again..
    ask_user = input("Do you wanna play again? Y or N: ").upper()
    if ask_user != "Y" and ask_user != "N":
        ask_user = input("Do you wanna play again? Y or N: ").upper()
    
    if ask_user == "Y":
        result = False
    else:
        result = True
    
print("Thank you for playing..!!!")
```
