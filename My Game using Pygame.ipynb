# Game with one press - one move type

```python
import pygame , sys , os
from pygame.locals import *

#Set attributes
x_axis = 0    #left,right
y_axis = 0    #up,down
screen = 0

#Creating func to quit
def quitgame():
    pygame.quit()
    sys.exit(0)

#Creating func for Event loop
def eventloop(some_event):
    global x_axis,y_axis,screen,colour
    for event in some_event:
        if event.type == QUIT:
            quit()
        else:
            if event.type == pygame.KEYDOWN:
                if event.key == pygame.K_LEFT or event.key == pygame.K_a:
                    x_axis -= 5
                    print("Moving left")
                elif event.key == pygame.K_RIGHT or event.key == pygame.K_d:
                    x_axis += 5
                    print("Moving right")
                elif event.key == pygame.K_UP or event.key == pygame.K_w:
                    y_axis -= 5
                    print("Moving up")
                elif event.key == pygame.K_DOWN or event.key == pygame.K_s:
                    y_axis += 5
                    print("Moving down")
                elif event.key == pygame.K_ESCAPE:
                    quitgame()

                else:
                    print("Invalid key pressed ",event.key)
    screen.fill(colour)
    pygame.draw.rect(screen,(0,0,0),(x_axis,y_axis,40,5))   #screen,(colour),(x,y,length,thick)
    pygame.draw.line(screen,(204,0,0),(x_axis,y_axis),(0,0),3)   #screen,(colour),(x1,y1),(x2,y2),thick
    pygame.display.update()
            
#Creating func main
def main():
    
    global screen,colour
    
    #Initialize pygame
    pygame.init()
    
    screen_width = 800
    screen_height = 500
    
    colour = [255,255,102] #Blue colour
    
    window = pygame.display.set_mode((screen_width, screen_height))
    pygame.display.set_caption('My Game')
                                     
    screen = pygame.display.get_surface()
    
    pygame.display.flip()
    pygame.display.update()
    
    while True:
        eventloop(pygame.event.get())
        #print(pygame.event.get())


main()
```

```python
import pygame , sys , os
from pygame.locals import *

#Set attributes
x_axis = 0    #left,right
y_axis = 490    #up,down
screen = 0
velocity = 1
widthOfRect = 60
heightOfRect = 7
screen_width = 900  #Max = 1366
screen_height = 500 #Max = 768
isJump = False
jumpCount = 10

#Creating func to quit
def quitgame():
    pygame.quit()
    sys.exit(0)

#Creating func for Event loop
def eventloop(some_event):
    global x_axis,y_axis,screen,colour,velocity,widthOfRect,screen_width,heightOfRect,screen_height,isJump,jumpCount
    for event in some_event:
        if event.type == pygame.QUIT:
            quitgame()
    
    keys_pressed = pygame.key.get_pressed()
    
    if keys_pressed[pygame.K_ESCAPE]:
        quitgame()
    
    if (keys_pressed[pygame.K_LEFT] or keys_pressed[pygame.K_a]) and x_axis > velocity:
        x_axis -= velocity
        
    if (keys_pressed[pygame.K_RIGHT] or keys_pressed[pygame.K_d]) and x_axis < screen_width - widthOfRect - velocity:
        x_axis += velocity
    
    if not(isJump):
        if (keys_pressed[pygame.K_UP] or keys_pressed[pygame.K_w]) and y_axis > velocity:
            y_axis -= velocity

        if (keys_pressed[pygame.K_DOWN] or keys_pressed[pygame.K_s]) and y_axis < screen_height - heightOfRect - velocity :
            y_axis += velocity

        if keys_pressed[pygame.K_SPACE]:
            isJump = True
    else:
        if jumpCount >= -10:
            val = 1
            if jumpCount < 0:
                val = -1
            y_axis -= (jumpCount ** 2) * 0.5 * val   #To make obj jump and return down
            jumpCount -= 1
        
        else:
            isJump = False
            jumpCount = 10
    
    screen.fill(colour)
    pygame.draw.rect(screen,(0,0,0),(x_axis,y_axis,widthOfRect,heightOfRect))   #screen,(colour),(x,y,length,thick)
    pygame.draw.line(screen,(204,0,0),(0,0),(x_axis,y_axis),3)   #screen,(colour),(x1,y1),(x2,y2),thick
    pygame.display.update()
            
#Creating func main
def main():
    
    global screen,colour,screen_width,screen_height
    
    #Initialize pygame
    pygame.init()
    
    colour = [255,255,102] #Blue colour
    
    window = pygame.display.set_mode((screen_width, screen_height))
    pygame.display.set_caption('My Game')
                                     
    screen = pygame.display.get_surface()
    
    pygame.display.flip()
    pygame.display.update()
    
    while True:
        eventloop(pygame.event.get())
        #print(pygame.event.get())


main()
```
